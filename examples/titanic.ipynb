### Simple Subgroup Discovery Example with Titanic data

```python
import pysubgroup as ps
from pysubgroup.datasets import get_titanic_data

# this gives you a pandas DataFrame
data = get_titanic_data() 

 # We want to find subgroups that describe people who survived
target = ps.BinaryTarget("Survived", True)

# We define, which basic conditions we want to consider. 
# Here: automatically generated from all attributes that are not the target, name or Ticket
searchspace = ps.create_selectors(data, ignore=["Survived", "Name", "Ticket"])

print(searchspace)

# Define the parameters of the search
task = ps.SubgroupDiscoveryTask(
    data, target, searchspace, result_set_size=5, depth=2, qf=ps.StandardQF(0.5)
)

# Run the algorithm
result = ps.BeamSearch().execute(task)

# Print the result
result.to_dataframe()
```
